```python
import os
import gym
from stable_baselines3 import PPO, DQN
from stable_baselines3.common.vec_env import DummyVecEnv
from stable_baselines3.common.evaluation import evaluate_policy
# additional methods
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```python
environment_name='CartPole-v0'
env = gym.make(environment_name)
```

```python
episodes = 10

for episode in range(1, episodes+1):
  score=0
  state = env.reset()
  done=False

  while not done:
    action = env.action_space.sample()
    # n_state is next state
    state, reward, info, done, _ = env.step(action)
    score += reward

  print('Episode {} Score{}'.format(episode, score))

  env.close()
```

for different algorithims for model free RL algorithims, go to https://spinningup.openai.com/en/latest/spinningup/rl_intro2.html

```python
log_path = os.path.join('Training', 'Logs')
```

```python
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])
model = PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
model.learn(total_timesteps=20000)
```

```python
PPO_path = os.path.join('Training', 'Saved Models', 'PPO_model')
```

```python
model.save(PPO_path)
```

Evaluation and Performance

```python
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

Testing the Model

```python
episodes = 10

for episode in range(1, episodes+1):
    score = 0
    obs = env.reset()
    done = False

    while not done:
        env.render()
#       all the training/learning for predicting the actions
        action, _ = model.predict(obs)
        obs, reward, done, info = env.step(action)
        score += reward

    print('Episode {} Score{}'.format(episode, score))

env.close()
```

```python
training_log_path = os.path.join(log_path, 'PPO_1')
training_log_path
save_path= os.path.join('Training', 'Saved Models')
```

```python
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=200, verbose=1)
eval_callback = EvalCallback(env,
                             callback_on_new_best=stop_callback,
                             eval_freq = 10000,
                             best_model_save_path=save_path,
                             verbose=1)
```

```python
model=PPO('MlpPolicy', env, verbose=1, tensorboard_log=log_path)
model.learn(total_timesteps=200000, callback=eval_callback)
```
